# 车辆MPG  数据分析

* Kaggle https://www.kaggle.com/uciml/autompg-dataset
* MPG = MILE PER GALLON = 每加伦汽油能距的英里数
* MPG 为标签类
* cylinders   displacement   horsepower	 weight	acceleration  model year	origin	car name 为特征列

```python
import pandas as pd
import matplotlib.pyplot as plt

#给数据添加标签
columns = ["mpg", "cylinders", "displacement", "horsepower", 
           "weight", "acceleration", "model year", "origin", "car name"] 
cars = pd.read_table("Data/1.auto-mpg.data", delim_whitespace=True, names=columns)
cars.head(5)
```

# 数据探索

```python
#创建一张图
fig = plt.figure() 
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
#绘制散点图
cars.plot("weight", "mpg", kind='scatter', ax=ax1) 
cars.plot("acceleration", "mpg", kind='scatter', ax=ax2)
plt.show()
```

# 建模 
* 线性回归

```python
import sklearn
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(cars[["weight"]], cars["mpg"])
```

```python
import sklearn
from sklearn.linear_model import LinearRegression

lr = LinearRegression(fit_intercept=True)
lr.fit(cars[["weight"]], cars["mpg"])
predictions = lr.predict(cars[["weight"]])

print(predictions[0:5])
print(cars["mpg"][0:5])
```

```python
plt.scatter(cars["weight"], cars["mpg"], c='red')
plt.scatter(cars["weight"], predictions, c='blue')
plt.show()
```

# 模型评估
* MSE （Mean Squared Error）叫做均方误差
* Root Mean Squard Error（RMSE）均方根误差

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(cars["mpg"], predictions)
mse
```

```python
mse = mean_squared_error(cars["mpg"], predictions)
rmse = mse ** (0.5)
rmse
```
